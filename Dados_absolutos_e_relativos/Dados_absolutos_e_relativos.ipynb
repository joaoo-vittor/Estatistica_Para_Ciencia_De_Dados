```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dados = {'empregos': ['Administrador_banco_dados', 'Programador', 'Arquiteto_redes'],
         'nova_jersey': [97350, 82080, 112840],
         'florida': [77140, 71540, 62310]}
```

```python
dataset = pd.DataFrame(dados)
dataset
```

```python
dataset['nova_jersey'].sum(), dataset['florida'].sum()
```

```python
dataset['%_nova_jersey'] = (dataset['nova_jersey'] / dataset['nova_jersey'].sum()) * 100
```

```python
dataset['%_florida'] = (dataset['florida'] / dataset['florida'].sum()) * 100
```

```python
data = {'matriculas_marco': [70, 50, 47, 23],
        'matriculas_novembro': [65, 48, 40, 22]}
```

```python
dataset_evasao = pd.DataFrame(data, index=np.arange(1,5))
```

```python
dataset_evasao
```

```python
aux = dataset_evasao['matriculas_marco'].sub(dataset_evasao['matriculas_novembro'], axis=0)
aux
```

```python
dataset_evasao['taxa_evasao'] = aux.div(dataset_evasao['matriculas_marco'], axis=0) * 100
```

```python
((dataset_evasao['matriculas_marco'].sum() - dataset_evasao['matriculas_novembro'].sum()) / dataset_evasao['matriculas_marco'].sum()) * 100
```
